```
import tensorflow as tf

tf.logging.set_verbosity(tf.logging.ERROR)

def eval_confusion_matrix(labels, predictions):
    with tf.variable_scope("eval_confusion_matrix"):
        con_matrix = tf.confusion_matrix(labels=labels, predictions=predictions, num_classes=10)

        con_matrix_sum = tf.Variable(tf.zeros(shape=(10,10), dtype=tf.int32),
                                            trainable=False,
                                            name="confusion_matrix_result",
                                            collections=[tf.GraphKeys.LOCAL_VARIABLES])

        update_op = tf.assign_add(con_matrix_sum, con_matrix)
        return tf.convert_to_tensor(con_matrix_sum), update_op

def cnn_model_fn(features, labels, mode):
    image_size = 96
    net = tf.reshape(features["x"], [-1, 96, 96, 3])
    # layer_input = input_layer
    for _ in range(layers_count):

        net = tf.layers.conv2d( #23
            inputs=net,
            filters=kernels_count,
            kernel_size=[kernels_size, kernels_size],  
            padding="same", ## same baes mishe size khoroji convalve ba size vorodi yeki bashe
            activation=tf.nn.relu)
        if pooling_method == "min":
            net = tf.multiply(net,-1)
            net = tf.layers.max_pooling2d(inputs=net, pool_size=[2, 2], strides=2) #16
            net = tf.multiply(net,-1)
        elif pooling_method == "max":
            net = tf.layers.max_pooling2d(inputs=net, pool_size=[2, 2], strides=2) #16
        elif pooling_method == "mean":
            net = tf.layers.average_pooling2d(inputs=net, pool_size=[2, 2], strides=2) #16
        if keep_prob < 1:
            net = tf.layers.dropout(
            inputs=net, rate=1-keep_prob, training=mode == tf.estimator.ModeKeys.TRAIN)
        # layer_input = net

    pool_flat = tf.reshape(net, [-1, int(image_size / 2**layers_count) * int(image_size / 2**layers_count)  * kernels_count])
    # # Logits Layer
    logits = tf.layers.dense(inputs=pool_flat, units=10)
    #logits = tf.layers.dense(inputs=dense2, units=10)
    predictions = {
        # Generate predictions (for PREDICT and EVAL mode)
        "classes": tf.argmax(input=logits, axis=1),
        # Add `softmax_tensor` to the graph. It is used for PREDICT and by the
        "probabilities": tf.nn.softmax(logits, name="softmax_tensor")
    }
    if mode == tf.estimator.ModeKeys.PREDICT:
        return tf.estimator.EstimatorSpec(mode=mode, predictions=predictions)

    # Calculate Loss (for both TRAIN and EVAL modes)
    loss = tf.losses.sparse_softmax_cross_entropy(labels=labels, logits=logits)
    accuracy = tf.metrics.accuracy(labels=labels, predictions=predictions["classes"])
    if mode == tf.estimator.ModeKeys.TRAIN:
        optimizer = tf.train.AdamOptimizer()
        train_op = optimizer.minimize(
            loss=loss,
            global_step=tf.train.get_global_step())
        logging_hook = tf.train.LoggingTensorHook({"loss" : loss, 
            "accuracy" : accuracy[1]}, every_n_iter=100)
        return tf.estimator.EstimatorSpec(mode=mode, loss=loss, train_op=train_op, training_hooks=[logging_hook])

    # Add evaluation metrics (for EVAL mode)
    eval_metric_ops = {
        "accuracy": tf.metrics.accuracy(
            labels=labels, predictions=predictions["classes"])
    }
    if mode == tf.estimator.ModeKeys.EVAL:
          return tf.estimator.EstimatorSpec(
          mode=mode,
          loss=loss,
          eval_metric_ops={'accuracy': accuracy, "conf_matrix": eval_confusion_matrix(labels, tf.argmax(input=logits, axis=1))},
          evaluation_hooks=None)
    return tf.estimator.EstimatorSpec(
        mode=mode, loss=loss, eval_metric_ops=eval_metric_ops)



from keras.datasets import mnist
import numpy as np
import read_data

# path to the directory with the data
DATA_DIR = './data'

# url of the binary data
DATA_URL = 'http://ai.stanford.edu/~acoates/stl10/stl10_binary.tar.gz'

# path to the binary train file with image data
DATA_PATH = './data/stl10_binary/train_X.bin'

# path to the binary train file with labels
LABEL_PATH = './data/stl10_binary/train_y.bin'

# Load training and eval data
images = read_data.read_all_images(DATA_PATH)
print(images.shape)

labels = read_data.read_labels(LABEL_PATH)
print(labels.shape)
data_count = 5000
train_rate = .8
train_index = int(data_count * train_rate)
train_data = images[:train_index].astype(np.float32)
train_labels = labels[:train_index].astype(np.int32) - 1  # not required

eval_data = images[train_index: data_count].astype(np.float32)
eval_labels = labels[ train_index: data_count].astype(np.int32) - 1  # not required

# tensors_to_log = {"probabilities": "softmax_tensor"}

# logging_hook = tf.train.LoggingTensorHook(
#     tensors=tensors_to_log, every_n_iter=50)

train_input_fn = tf.estimator.inputs.numpy_input_fn(
    x={"x": train_data},
    y=train_labels,
    batch_size=64,
    num_epochs=1,
    shuffle=True)

train_input_test_fn = tf.estimator.inputs.numpy_input_fn(
    x={"x": train_data},
    y=train_labels,
    batch_size=64,# maybe gooood 
    num_epochs=1,
    shuffle=True)

eval_input_fn = tf.estimator.inputs.numpy_input_fn(
            x={"x": eval_data},
            y=eval_labels,
            batch_size=64,
            num_epochs=1,
            shuffle=True)
def train_and_summery():
    t_losses = []
    e_losses = []
    t_accs = []
    e_accs = []
    mnist_classifier = tf.estimator.Estimator(
        model_fn=cnn_model_fn)#,model_dir="./model_first")

    for epoch in range(EPOCHS):
        
        mnist_classifier.train(input_fn=train_input_fn)#, hooks=[acc_hook])
        
        eval_results = mnist_classifier.evaluate(input_fn=eval_input_fn)
        train_results = mnist_classifier.evaluate(input_fn=train_input_test_fn)
        # train_results = eval_results
        t_losses.append(train_results['loss'])
        t_accs.append(train_results['accuracy'])
        e_losses.append(eval_results['loss'])
        e_accs.append(eval_results['accuracy'])
        print(".", end ="")
        # print("epoch: ", epoch, "train acc:",train_results['accuracy'] ,"eval acc", eval_results['accuracy'])
        # print(eval_results)
    print("epoch: ", epoch, "train acc:",train_results['accuracy'] ,"eval acc", eval_results['accuracy'])
    print("train conf matrix:", train_results['conf_matrix'])
    print("eval conf matrix:", eval_results['conf_matrix'])
    plt.plot(t_losses, label='t_losses', alpha=0.6)
    plt.plot(e_losses, label='e_losses', alpha=0.6)
    plt.title("Losses")
    plt.show()
    plt.plot(t_accs, label='t_accs', alpha=0.6)
    plt.plot(e_accs, label='e_accs', alpha=0.6)
    plt.title("Accuracy")
    plt.show()
    plt.close()

EPOCHS = 20
kernels_count = 16
kernels_size = 2
layers_count = 4
pooling_method = "mean"
keep_prob = 1

for layers_count in [2, 3, 4]:
    print("layers_count:", layers_count)
    train_and_summery()

layers_count = 4
for kernels_count in [8, 32, 64]:
    print("kernels_count:", kernels_count)
    train_and_summery()

for kernels_size in [2, 3, 4]:
    print("kernels_size:", kernels_size)
    train_and_summery()
kernels_size = 2

for pooling_method in ["min", "max", "mean"]:
    print("pooling_method:", pooling_method)
    train_and_summery()
pooling_method = "mean"
EPOCHS = 30
for keep_prob in [1, .4, .8]:
    print("keep_prob:", keep_prob)
    train_and_summery()
```
